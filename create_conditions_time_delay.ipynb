```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
time_delays=[]
trialPerCondition=12

audio_delays=[-165,-99,-66,-33,0,33,66,99,165]
visual_delays=[-50,-33,-16,0,16,33,50]

audio_delays=[-10,-6,-4,-2,0,2,4,6,10]
visual_delays=[-4,-2,-1,0,1,2,4]

lenAudio=len(audio_delays)

# create all possible combinations of audio and visual delays
audio_visual_delays=[]
for i in range(len(audio_delays)):
    for j in range(len(visual_delays)):
        audio_visual_delays.append((audio_delays[i],visual_delays[j]))

audio_visual_delays=np.tile(audio_visual_delays,trialPerCondition)

# create a matrix of audio_visual_delays
audio_visual_delays=np.array(audio_visual_delays)
sh=audio_visual_delays.shape
dims=(sh[0]*sh[1])//2
#repmat it to create a matrix of mx2
audio_visual_delays=audio_visual_delays.reshape(dims,2)

# shuffle the matrix but keep the audio and visual delays together
np.random.shuffle(audio_visual_delays)
trialNum=len(audio_visual_delays)
# plot the audio visual delays
plt.figure()
plt.plot(audio_visual_delays[:,0],audio_visual_delays[:,1],'o')
plt.xlabel('Audio delays')
plt.ylabel('Visual delays')
plt.title('Audio Visual delays')
plt.show()
```

```python
# Not the aim is to create a uniform random array of time delays between 300 and 900 ms
onsetFlashTimes=[]
onsetFlashTimes=np.random.uniform(300,900,trialNum)
minTime=min(onsetFlashTimes)
maxTime=max(onsetFlashTimes)
plt.figure()
plt.hist(onsetFlashTimes,bins=10)
plt.xlabel('Time delays')
plt.ylabel('Frequency')
plt.title('Time delays histogram')
plt.show()
```
